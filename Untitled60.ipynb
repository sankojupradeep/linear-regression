```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mart=pd.read_excel("C:\\Users\\Hello\\Downloads\\supermarket_sales.xlsx")
mart=mart[['Gender','Payment','Unit price','Quantity','Total','gross income']]
mart.head(10)
```

```python
sns.histplot(data=mart,x="Total",y="gross income")
```

```python
sns.kdeplot(data=mart,x="Total",hue="Payment",multiple="stack")
```

```python
sns.kdeplot(data=mart,x="Total",)
```

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=mart,y="Unit price",x="gross income",hue="Gender",fill=True)
sns.rugplot(data=mart,x='Unit price')
```

```python
sns.ecdfplot(data=mart,x="gross income",complementary=False,stat='count')
```

```python
sns.ecdfplot(data=mart,x="gross income",complementary=True)
```

```python
sns.displot(data=mart,x="Total",kde=True,rug=True,rug_kws={'height':0.05})
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length",kde=True)
```

```python
l=sns.jointplot(data=iris,x="sepal_length",y="petal_length",joint_kws=dict(marker='+'),color='yellow',marginal_kws=dict(color='red',kde=True),marginal_ticks=True)
l.plot_joint(sns.kdeplot)
```

```python
sns.pairplot(iris,diag_kws=dict(kde=True),plot_kws=dict(color="green",marker=10,s=5))
```

```python
l=sns.pairplot(iris,plot_kws=dict(color='red',marker=10),diag_kws=dict(kde=True))
l.map_upper(sns.kdeplot)
l.map_lower(sns.kdeplot)
```
